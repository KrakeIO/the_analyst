```python
import pandas as pd
import numpy as np
pd.options.display.float_format = '{:.2f}'.format


df = pd.read_csv('training_data/training_data.csv')
```

```python
target = df['recovered'] 
```

```python
raw_inputs = df[[
    'industry_id','company_id', 
    'pb_ratio', 'pc_ratio', 
    'pe_ratio','ps_ratio', 
    'discount_percentile', 'institution_held',
    'insider_held','return_on_assets',
    'return_on_equity','total_debt_equity',
    'current_ratio', 'market_cap',
    'transaction_id'
]]

print(raw_inputs)

clean_inputs = df[[
    'industry_id','company_id', 
    'pb_ratio', 'pc_ratio', 
    'pe_ratio','ps_ratio', 
    'discount_percentile', 'institution_held',
    'insider_held','return_on_assets',
    'return_on_equity','total_debt_equity',
    'current_ratio', 'market_cap'
]]

print(clean_inputs)
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
```

```python
# nb.predict([clean_inputs.iloc[200]])
```

```python
print raw_inputs.iloc[200]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(clean_inputs, target, random_state=1)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
nb.fit(X_train, y_train)
```

```python
predictions = nb.predict(X_test)
correct_predictions_size = sum(predictions == y_test)

incorrect_predictions_size = y_test.size - correct_predictions_size  # (there are 9,092 tweets in the csv)
print('Correct predictions: ' + str(correct_predictions_size))
print('Incorrect predictions: ' + str(incorrect_predictions_size))
print('Percent correct: ' + str(100.0 * correct_predictions_size / (correct_predictions_size + incorrect_predictions_size)))
```
